```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, recall_score
from sklearn.decomposition import PCA
```

```python
# 1. Load Kalboard360 Dataset
data = pd.read_csv('./Kalboard360.csv')

# 2. Data Preprocessing
# Change column names to more sophisticated ones
data.columns = ["Gender", "Nationality", "Place_of_Birth", "Grade_Level", "StageID", "GradeId", "Section_ID", "Topic", 
                "Semester", "Relation", "Raised_Hand", "Visited_Resources", "Viewing_Announcements", 
                "Discussion_Groups", "Parent_Satisfaction", "Student_Absence", "Class"]

# Convert categorical variables to numeric values
for column in data.select_dtypes(include=['object']).columns:
    data[column] = LabelEncoder().fit_transform(data[column])

# Handling missing values (if any)
data.fillna(data.mean(), inplace=True)
```

```python
# 3. Plot Graphs for Each Feature
for col in data.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Split Data
X = data.drop("Class", axis=1)
y = data["Class"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Helper function to evaluate models
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="weighted")
    recall = recall_score(y_true, y_pred, average="weighted")
    return accuracy, precision, recall
```

```python
# 4. Decision Tree, ANN, and Naive Bayes
# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_accuracy, dt_precision, dt_recall = evaluate_model(y_test, dt_pred)

# Plot Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(dt_model, feature_names=X.columns, class_names=["Low", "Middle", "High"], filled=True)
plt.show()
```

```python
# ANN
ann_model = MLPClassifier(max_iter=300, random_state=42)
ann_model.fit(X_train, y_train)
ann_pred = ann_model.predict(X_test)
ann_accuracy, ann_precision, ann_recall = evaluate_model(y_test, ann_pred)
```

```python
# Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)
nb_accuracy, nb_precision, nb_recall = evaluate_model(y_test, nb_pred)
```

```python
# Display Confusion Matrices
models = [("Decision Tree", dt_pred), ("ANN", ann_pred), ("Naive Bayes", nb_pred)]
for model_name, y_pred in models:
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

# Display Precision, Accuracy, and Recall in Tabular Format
results = pd.DataFrame({
    "Model": ["Decision Tree", "ANN", "Naive Bayes"],
    "Accuracy": [dt_accuracy, ann_accuracy, nb_accuracy],
    "Precision": [dt_precision, ann_precision, nb_precision],
    "Recall": [dt_recall, ann_recall, nb_recall]
})
print(results)
```

```python
# 5. Clustering Ensemble (KMeans Example)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_train)
cluster_labels = kmeans.predict(X_test)
cluster_accuracy, cluster_precision, cluster_recall = evaluate_model(y_test, cluster_labels)

# Confusion Matrix for Clustering Ensemble
cm = confusion_matrix(y_test, cluster_labels)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Clustering Ensemble')
plt.show()
```

```python
# 6. Apply PCA
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(pca.n_components_)], index=X.columns)

# Calculate the overall importance of each original feature by summing the absolute values across the top 5 components
feature_importance = loadings.iloc[:, :5].abs().sum(axis=1)

# Select the top 5 most important original features
top_features = feature_importance.sort_values(ascending=False).head(5)
print("Top 5 features selected by PCA:")
print(top_features)

# Plot Top 5 PCA Features
for i in range(5):
    plt.figure()
    plt.scatter(X_train_pca[:, i], y_train, alpha=0.5, label=f"Feature {i+1}")
    plt.title(f"PCA Feature {i+1} Impact on Class")
    plt.legend()
    plt.show()
```

```python
# 7. Decision Tree, ANN, and Naive Bayes with PCA Features
# Decision Tree with PCA
dt_model_pca = DecisionTreeClassifier()
dt_model_pca.fit(X_train_pca, y_train)
dt_pred_pca = dt_model_pca.predict(X_test_pca)
dt_accuracy_pca, dt_precision_pca, dt_recall_pca = evaluate_model(y_test, dt_pred_pca)

# Plot Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(dt_model_pca, feature_names=X.columns, class_names=["Low", "Middle", "High"], filled=True)
plt.show()
```

```python
# ANN with PCA
ann_model_pca = MLPClassifier(max_iter=300, random_state=42)
ann_model_pca.fit(X_train_pca, y_train)
ann_pred_pca = ann_model_pca.predict(X_test_pca)
ann_accuracy_pca, ann_precision_pca, ann_recall_pca = evaluate_model(y_test, ann_pred_pca)
```

```python
# Naive Bayes with PCA
nb_model_pca = GaussianNB()
nb_model_pca.fit(X_train_pca, y_train)
nb_pred_pca = nb_model_pca.predict(X_test_pca)
nb_accuracy_pca, nb_precision_pca, nb_recall_pca = evaluate_model(y_test, nb_pred_pca)
```

```python
models = [("Decision Tree", dt_pred_pca), ("ANN", ann_pred_pca), ("Naive Bayes", nb_pred_pca)]
for model_name, y_pred in models:
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()
    
# Display Results with PCA
results_pca = pd.DataFrame({
    "Model": ["Decision Tree (PCA)", "ANN (PCA)", "Naive Bayes (PCA)"],
    "Accuracy": [dt_accuracy_pca, ann_accuracy_pca, nb_accuracy_pca],
    "Precision": [dt_precision_pca, ann_precision_pca, nb_precision_pca],
    "Recall": [dt_recall_pca, ann_recall_pca, nb_recall_pca]
})
print(results_pca)
```

```python
# 8. Clustering Ensemble with PCA
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca.fit(X_train_pca)
cluster_labels_pca = kmeans_pca.predict(X_test_pca)
cluster_accuracy_pca, cluster_precision_pca, cluster_recall_pca = evaluate_model(y_test, cluster_labels_pca)

# 9. Comparison Plot for PCA vs Non-PCA Accuracy
plt.figure(figsize=(10, 5))
plt.bar(["Decision Tree", "ANN", "Naive Bayes"], results["Accuracy"], alpha=0.5, label="Without PCA")
plt.bar(["Decision Tree (PCA)", "ANN (PCA)", "Naive Bayes (PCA)"], results_pca["Accuracy"], alpha=0.5, label="With PCA")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Comparison of Accuracy with and without PCA")
plt.show()
```

```python
# 10. Testing New Record
new_record = np.array([[1, 0, 2, 1, 30, 12, 0, 4, 70, 90, 5, 1, 5, 0, 3, 1]]).reshape(1, -1)
new_record_pca = pca.transform(new_record)

# Example prediction with PCA-enabled models
print("DT Prediction (PCA):", dt_model_pca.predict(new_record_pca))
print("ANN Prediction (PCA):", ann_model_pca.predict(new_record_pca))
print("Naive Bayes Prediction (PCA):", nb_model_pca.predict(new_record_pca))
```
