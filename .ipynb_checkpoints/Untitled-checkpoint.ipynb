```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv('./Kalboard360.csv')  # Replace with the path of your dataset

# 1. Rename columns for better understanding (Optional: Based on the dataset)
data.columns = ['Gender', 'Nationality', 'PlaceOfBirth', 'EducationalStage', 
                'GradeLevel', 'SectionID', 'Topic', 'Semester', 'Relation', 
                'RaisedHands', 'VisitedResources', 'AnnouncementsView', 
                'Discussion', 'ParentAnsweringSurvey', 'ParentsSchoolSatisfaction', 
                'StudentAbsenceDays', 'Class']

# 2. Handle missing values (if any)
# data.fillna(data.median(), inplace=True)

# 3. Separate categorical and numerical features
categorical_cols = ['Gender', 'Nationality', 'PlaceOfBirth', 'EducationalStage', 
                    'GradeLevel', 'SectionID', 'Topic', 'Semester', 'Relation', 
                    'ParentAnsweringSurvey', 'ParentsSchoolSatisfaction', 
                    'StudentAbsenceDays']

numerical_cols = ['RaisedHands', 'VisitedResources', 'AnnouncementsView', 'Discussion']

# Target variable
y = data['Class']
```

```python
from sklearn.preprocessing import LabelEncoder

# Example for converting 'Nationality' column using LabelEncoder
le = LabelEncoder()
data['Nationality'] = le.fit_transform(data['Nationality'])
```

```python
# 4. Encode Categorical Variables
#    - Binary categorical variables with Label Encoding
#    - Multi-category variables with One-Hot Encoding

# Apply Label Encoding to binary categorical columns
binary_cols = ['Gender', 'ParentAnsweringSurvey', 'ParentsSchoolSatisfaction', 'Relation']
le = LabelEncoder()
for col in binary_cols:
    data[col] = le.fit_transform(data[col])

# Apply One-Hot Encoding to multi-class categorical variables
multi_class_cols = ['Nationality', 'PlaceOfBirth', 'EducationalStage', 'GradeLevel', 
                    'SectionID', 'Topic', 'Semester', 'StudentAbsenceDays']

# Perform One-Hot Encoding using pd.get_dummies()
data = pd.get_dummies(data, columns=multi_class_cols)
```

```python
# 5. Separate features and target after encoding
X = data.drop('Class', axis=1)

# 6. Standardization of only numerical columns
# Apply standardization only to numerical features
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

# Visualizations
# --------------
# Plot histograms for each feature (only for numerical features)
X[numerical_cols].hist(figsize=(15, 12))
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(X[numerical_cols].corr(), annot=False, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Algorithms Before PCA
# ---------------------
# Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {title}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    print(f"{title} Classification Report:\n", classification_report(y_true, y_pred))
    
plot_confusion_matrix(y_test, y_pred_dt, 'Decision Tree Before PCA')


```

```python
# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {title}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    print(f"{title} Classification Report:\n", classification_report(y_true, y_pred))
    
plot_confusion_matrix(y_test, y_pred_nb, 'Naive Bayes Before PCA')


```

```python
# Artificial Neural Network
ann = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=200, random_state=42)
ann.fit(X_train, y_train)
y_pred_ann = ann.predict(X_test)

# Confusion Matrices
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {title}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    print(f"{title} Classification Report:\n", classification_report(y_true, y_pred))

plot_confusion_matrix(y_test, y_pred_ann, 'ANN Before PCA')

    
# Clustering Ensemble Before PCA
# ------------------------------
kmeans_before = KMeans(n_clusters=len(np.unique(y)), random_state=42)
kmeans_before.fit(X)
data['Cluster'] = kmeans_before.labels_
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Encode the target variable 'Class' using LabelEncoder
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)  # y_encoded should have 480 samples

# Fit KMeans and ensure it has 480 labels
kmeans_before = KMeans(n_clusters=3, random_state=0).fit(X)  # X is your input data
cluster_labels_before = kmeans_before.labels_

# Function to map clusters to labels based on majority voting
def map_clusters_to_labels(cluster_labels, true_labels):
    label_mapping = {}
    for cluster in np.unique(cluster_labels):
        mask = cluster_labels == cluster
        majority_label = np.bincount(true_labels[mask]).argmax()
        label_mapping[cluster] = majority_label
    mapped_labels = np.vectorize(label_mapping.get)(cluster_labels)
    return mapped_labels

# Apply the mapping function
cluster_labels_before_mapped = map_clusters_to_labels(cluster_labels_before, y_encoded)

# Plot confusion matrix for clustering ensemble before PCA
def plot_confusion_matrix(true_labels, predicted_labels, title):
    cm = confusion_matrix(true_labels, predicted_labels)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

# Plot the confusion matrix
plot_confusion_matrix(y_encoded, cluster_labels_before_mapped, 'Clustering Ensemble Before PCA')
```

```python
print(len(y_encoded), len(cluster_labels_before))
print(len(kmeans_before.labels_))
```

```python
# Feature Selection using PCA
# ---------------------------
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)
X_pca_train, X_pca_test = train_test_split(X_pca, test_size=0.3, random_state=42)

explained_variance = pca.explained_variance_ratio_

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(range(1, 6), explained_variance, alpha=0.7, color='b', align='center')
plt.xticks(range(1, 6), [f'Feature {i}' for i in range(1, 6)])
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.title('Explained Variance by Top 5 Principal Components')
plt.show()
```

```python
# Algorithms After PCA
# --------------------
# Decision Tree
dt_pca = DecisionTreeClassifier(random_state=42)
dt_pca.fit(X_pca_train, y_train)
y_pred_dt_pca = dt_pca.predict(X_pca_test)
plot_confusion_matrix(y_test, y_pred_dt_pca, 'Decision Tree After PCA')
```

```python
# Naive Bayes
nb_pca = GaussianNB()
nb_pca.fit(X_pca_train, y_train)
y_pred_nb_pca = nb_pca.predict(X_pca_test)
plot_confusion_matrix(y_test, y_pred_nb_pca, 'Naive Bayes After PCA')

```

```python
# Artificial Neural Network
ann_pca = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=200, random_state=42)
ann_pca.fit(X_pca_train, y_train)
y_pred_ann_pca = ann_pca.predict(X_pca_test)
plot_confusion_matrix(y_test, y_pred_ann_pca, 'ANN After PCA')


```

```python

# Clustering Ensemble After PCA
kmeans_after = KMeans(n_clusters=len(np.unique(y)), random_state=42)
kmeans_after.fit(X_pca)
data['Cluster_PCA'] = kmeans_after.labels_

# Mapping clusters to labels for the clustering ensemble after PCA
cluster_labels_after = map_clusters_to_labels(kmeans_after.labels_, y_encoded)  # Use numeric labels
# Plot confusion matrix for clustering ensemble after PCA
plot_confusion_matrix(y_encoded, cluster_labels_after, 'Clustering Ensemble After PCA')


# Accuracy Comparison Before and After PCA
# ----------------------------------------
models = ['Decision Tree', 'Naive Bayes', 'ANN']
accuracy_before = [accuracy_score(y_test, y_pred_dt), accuracy_score(y_test, y_pred_nb), accuracy_score(y_test, y_pred_ann)]
accuracy_after = [accuracy_score(y_test, y_pred_dt_pca), accuracy_score(y_test, y_pred_nb_pca), accuracy_score(y_test, y_pred_ann_pca)]

# Plotting Accuracy Comparison
plt.figure(figsize=(10, 6))
x = np.arange(len(models))
width = 0.35
plt.bar(x - width/2, accuracy_before, width, label='Before PCA')
plt.bar(x + width/2, accuracy_after, width, label='After PCA')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Before and After PCA')
plt.xticks(x, models)
plt.legend()
plt.show()
```
