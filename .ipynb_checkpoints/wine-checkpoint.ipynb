```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv("./wine.csv") 
print(data.columns)
print(data.head)

# Check for missing values
data.isnull().sum()

# Fill or drop missing values if any 
data = data.dropna()  

# Separate features and target variable
X = data.drop(columns=['type'])  
y = data['type']

# Standardize the features for DBSCAN and PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert target to numeric labels
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(data['type'])  # Convert 'type' column to numeric labels
X = data.drop(columns=['type'])
```

```python
import matplotlib.pyplot as plt

# Plot individual histograms for each feature
features = X.columns
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features, 1):
    plt.subplot((len(features) + 1) // 3, 3, i)  # Adjust layout for clarity
    plt.hist(X[feature], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.tight_layout()

plt.suptitle('Histograms of Each Feature in the Dataset', fontsize=16)
plt.subplots_adjust(top=0.9)  # Adjust title position
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

# Assume `X` is your dataset after preprocessing and PCA (if applied).

# Step 1: k-distance plot to find optimal eps
k = 10  # set minPts around 10 based on the rule of thumb
nearest_neighbors = NearestNeighbors(n_neighbors=k)
neighbors = nearest_neighbors.fit(X)
distances, indices = neighbors.kneighbors(X)

# Sort distances for k-distance plot
distances = np.sort(distances[:, k-1], axis=0)
plt.plot(distances)
plt.ylabel("k-distance")
plt.xlabel("Points sorted by distance")
plt.title("k-distance plot for DBSCAN")
plt.show()

# Step 2: Run DBSCAN with selected eps and minPts
dbscan = DBSCAN(eps=0.13, min_samples=k)
dbscan.fit(X)

# Count noise points
n_noise = list(dbscan.labels_).count(-1)
print("Number of noise points:", n_noise)
```

```python
# Initialize DBSCAN
dbscan_no_pca = DBSCAN(eps=0.13, min_samples=26)
y_pred_no_pca = dbscan_no_pca.fit_predict(X_scaled)

# Visualize clusters (use PCA for visualization, without fitting it fully)
pca_vis = PCA(n_components=2)
X_vis = pca_vis.fit_transform(X_scaled)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_vis[:, 0], y=X_vis[:, 1], hue=y_pred_no_pca, palette='viridis', s=50)
plt.title('DBSCAN Clusters without PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Calculate confusion matrix, precision, recall, and accuracy
cm_no_pca = confusion_matrix(y, y_pred_no_pca)
precision_no_pca = precision_score(y, y_pred_no_pca, average='weighted', zero_division=1)
recall_no_pca = recall_score(y, y_pred_no_pca, average='weighted', zero_division=1)
accuracy_no_pca = accuracy_score(y, y_pred_no_pca)
```

```python
# Apply PCA for feature selection
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)



# Fit PCA for feature selection (assuming 5 components as in the previous example)
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)

# Plot explained variance for each principal component
plt.figure(figsize=(8, 6))
plt.bar(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_, color='teal', alpha=0.7)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Variance Explained by Each Principal Component')
plt.xticks(range(1, pca.n_components_ + 1))
plt.show()

# Apply DBSCAN on PCA-transformed data
dbscan_with_pca = DBSCAN(eps=0.13, min_samples=10)
y_pred_with_pca = dbscan_with_pca.fit_predict(X_pca)

# Visualize clusters post-PCA
X_vis_pca = pca_vis.fit_transform(X_pca)  # For 2D visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_vis_pca[:, 0], y=X_vis_pca[:, 1], hue=y_pred_with_pca, palette='viridis', s=50)
plt.title('DBSCAN Clusters with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Calculate confusion matrix, precision, recall, and accuracy after PCA
cm_with_pca = confusion_matrix(y, y_pred_with_pca)
precision_with_pca = precision_score(y, y_pred_with_pca, average='weighted', zero_division=1)
recall_with_pca = recall_score(y, y_pred_with_pca, average='weighted', zero_division=1)
accuracy_with_pca = accuracy_score(y, y_pred_with_pca)
```

```python
print("Without PCA:")
print("Precision:", precision_no_pca)
print("Recall:", recall_no_pca)
print("Accuracy:", accuracy_no_pca)
print("\nWith PCA:")
print("Precision:", precision_with_pca)
print("Recall:", recall_with_pca)
print("Accuracy:", accuracy_with_pca)

# Determine better model
if accuracy_with_pca > accuracy_no_pca:
    print("Model with PCA performs better.")
else:
    print("Model without PCA performs better.")
    
# Values for each metric
metrics = ['Precision', 'Recall', 'Accuracy']
no_pca_values = [precision_no_pca, recall_no_pca, accuracy_no_pca]
with_pca_values = [precision_with_pca, recall_with_pca, accuracy_with_pca]

# Plotting
x = np.arange(len(metrics))  # Label locations
width = 0.35  # Width of the bars

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, no_pca_values, width, label='Without PCA')
bars2 = ax.bar(x + width/2, with_pca_values, width, label='With PCA')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Model Performance with and without PCA')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

# Displaying values on top of each bar
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.01, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Function to display noise points and number of clusters
def dbscan_cluster_info(labels):
    # Count the number of clusters (excluding noise labeled as -1)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    # Count the number of noise points
    n_noise = list(labels).count(-1)
    
    # Display the results in a table format
    info = pd.DataFrame({
        'Metric': ['Number of Clusters', 'Number of Noise Points'],
        'Without PCA': [n_clusters if 'no_pca' in locals() else None, n_noise if 'no_pca' in locals() else None],
        'With PCA': [n_clusters if 'with_pca' in locals() else None, n_noise if 'with_pca' in locals() else None]
    })
    return info

# For DBSCAN without PCA
n_clusters_no_pca = len(set(y_pred_no_pca)) - (1 if -1 in y_pred_no_pca else 0)
n_noise_no_pca = list(y_pred_no_pca).count(-1)

# For DBSCAN with PCA
n_clusters_with_pca = len(set(y_pred_with_pca)) - (1 if -1 in y_pred_with_pca else 0)
n_noise_with_pca = list(y_pred_with_pca).count(-1)

# Display the table
results_df = pd.DataFrame({
    'Metric': ['Number of Clusters', 'Number of Noise Points'],
    'Without PCA': [n_clusters_no_pca, n_noise_no_pca],
    'With PCA': [n_clusters_with_pca, n_noise_with_pca]
})
print(results_df)
```

```python
# Plotting the results
fig, ax = plt.subplots(figsize=(8, 5))

# Setting up bar positions and width
metrics = results_df['Metric']
x = range(len(metrics))  
width = 0.35  

# Plotting bars for each result
bars1 = ax.bar([i - width/2 for i in x], results_df['Without PCA'], width, label='Without PCA')
bars2 = ax.bar([i + width/2 for i in x], results_df['With PCA'], width, label='With PCA')

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Metrics')
ax.set_ylabel('Counts')
ax.set_title('DBSCAN Clustering Results with and without PCA')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Displaying values on top of each bar for clarity
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 1, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
