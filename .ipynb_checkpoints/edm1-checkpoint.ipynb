```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.cluster import KMeans
from sklearn.feature_selection import mutual_info_classif
from sklearn import tree
import graphviz

# # Load the Kalboard360 dataset
# # Replace 'kalboard360.csv' with the actual path to your Kalboard360 dataset
# df = pd.read_csv('kalboard360.csv')

# # Display a few rows of the dataset to verify the structure
# print(df.head())

# Identify the target column ('Class') and separate it from the features
X = df.drop('Class', axis=1)
y = df['Class']

# Check for categorical columns and convert them using Label Encoding
categorical_columns = X.select_dtypes(include=['object']).columns
label_encoder = LabelEncoder()

for col in categorical_columns:
    X[col] = label_encoder.fit_transform(X[col])

# Ensure the target variable 'y' is also encoded if it is categorical
if y.dtype == 'object':
    y = label_encoder.fit_transform(y)

# Get the top N features based on Information Gain
# Select only top features
top_n = 5  # Top 5 features
mi_scores = mutual_info_classif(X, y)
mi_feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Information Gain': mi_scores
}).sort_values(by='Information Gain', ascending=False)
top_features = mi_feature_importance.head(top_n)['Feature'].values
X = X[top_features]

# Plot the feature importance based on Information Gain
plt.figure(figsize=(10, 6))
sns.barplot(x='Information Gain', y='Feature', data=mi_feature_importance.head(top_n))
plt.title('Top Features based on Information Gain')
plt.show()

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the data for ANN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

### Model 1: Decision Tree Classifier ###
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
y_pred_dt = dt_classifier.predict(X_test)

# Plot the Decision Tree using plot_tree
plt.figure(figsize=(20, 10))
plot_tree(dt_classifier, feature_names=top_features, class_names=['Low', 'Middle', 'High'], filled=True)
plt.title('Decision Tree Visualization')
plt.show()

# Alternatively, using graphviz to visualize the Decision Tree (if installed)
dot_data = tree.export_graphviz(dt_classifier, out_file=None, 
                                feature_names=top_features,  
                                class_names=['Low', 'Middle', 'High'],  
                                filled=True, rounded=True,  
                                special_characters=True)  
graph = graphviz.Source(dot_data)  
graph.render("Decision_Tree_Kalboard360")  # Saves the visualization to a file

### Model 2: Artificial Neural Network (ANN) ###
ann_classifier = MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000, random_state=42)
ann_classifier.fit(X_train_scaled, y_train)
y_pred_ann = ann_classifier.predict(X_test_scaled)

### Model 3: Naive Bayes Classifier ###
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred_nb = nb_classifier.predict(X_test)

# Evaluate each model
models = {
    'Decision Tree': y_pred_dt,
    'Artificial Neural Network (ANN)': y_pred_ann,
    'Naive Bayes': y_pred_nb
}

for model_name, y_pred in models.items():
    print(f"\n{model_name} Confusion Matrix")
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

    acc = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {acc*100:.2f}%")

### Model 4: Clustering Ensemble using KMeans ###
# Combine predictions for clustering
combined_predictions = np.array([y_pred_dt, y_pred_ann, y_pred_nb]).T

# Apply KMeans to cluster the predictions
kmeans = KMeans(n_clusters=len(np.unique(y_test)), random_state=42)
kmeans.fit(combined_predictions)
ensemble_pred = kmeans.labels_

# Confusion Matrix and Accuracy for Clustering Ensemble
cm_ensemble = confusion_matrix(y_test, ensemble_pred)
sns.heatmap(cm_ensemble, annot=True, fmt="d", cmap='Greens')
plt.title(f'Confusion Matrix - Clustering Ensemble')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

acc_ensemble = accuracy_score(y_test, ensemble_pred)
print(f"Clustering Ensemble Accuracy: {acc_ensemble*100:.2f}%")

# Additional Visualization: Feature Importance from Decision Tree
feature_importance = pd.Series(dt_classifier.feature_importances_, index=top_features).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance, y=feature_importance.index)
plt.title('Feature Importance from Decision Tree')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()

# Create a DataFrame to store the accuracy of each model
accuracy_results = {
    'Model': ['Decision Tree', 'Artificial Neural Network (ANN)', 'Naive Bayes', 'Clustering Ensemble'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_dt),
        accuracy_score(y_test, y_pred_ann),
        accuracy_score(y_test, y_pred_nb),
        accuracy_score(y_test, ensemble_pred)
    ]
}

accuracy_df = pd.DataFrame(accuracy_results)

# Summary of results and identification of the best model
best_model = accuracy_df.loc[accuracy_df['Accuracy'].idxmax()]

# Display the accuracy results as a table
print("\nModel Accuracy Comparison:")
print(accuracy_df)

# Display the accuracy results as a table
print("\nModel Accuracy Comparison:")
print(accuracy_df)

# Plot Model Accuracies
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=accuracy_df)
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limits from 0 to 1
plt.xticks(rotation=45)
plt.show()

# Print the best model summary
print("\nSummary of Model Performance:")
print(f"The best-performing model is: {best_model['Model']} with an accuracy of {best_model['Accuracy'] * 100:.2f}%.")
```
